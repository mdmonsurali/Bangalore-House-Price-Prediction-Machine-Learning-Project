## Detect Outlier

```python
# function to create histogram, Q-Q plot and boxplot

# for Q-Q plots
import scipy.stats as stats

def diagnostic_plots(df, variable):
    # function takes a dataframe (df) and
    # the variable of interest as arguments

    # define figure size
    plt.figure(figsize=(16, 4))

    # histogram
    plt.subplot(1, 3, 1)
    sns.distplot(df[variable], bins=30)
    plt.title('Histogram')

    # Q-Q plot
    plt.subplot(1, 3, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.ylabel('Variable quantiles')

    # boxplot
    plt.subplot(1, 3, 3)
    sns.boxplot(y=df[variable])
    plt.title('Boxplot')

    plt.show()
```

```python
num_var = ["bath","balcony","total_sqft_int","bhk","price"]
for var in num_var:
  print("******* {} *******".format(var))
  diagnostic_plots(df3, var)

  # here we observe outlier using histogram,, qq plot and boxplot 
```

```python
# location count is greter than 19 then we create column of that feature 
# then if this location present in location feature then set value 1 else 0 ( ohe hot encoding)
df10 = df9.copy()
for cat_var in location_gert_20:
  df10['location_'+cat_var]=np.where(df10['location']==cat_var, 1,0)
df10.shape
```

# Apply ML Model

```python
X = df11.drop("price", axis=1)
y = df11['price']
print('Shape of X = ', X.shape)
print('Shape of y = ', y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

print('Shape of X_train = ', X_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of X_test = ', X_test.shape)
print('Shape of y_test = ', y_test.shape)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train= sc.transform(X_train)
X_test = sc.transform(X_test)
```

# Model Building

```python
# impoer requiede packages
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

# **Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
#Check prediction Accuracy
lr_score = lr.score(X_test, y_test)

#Root mean Squre Error
lr_rmse = np.sqrt(mean_squared_error(y_test, lr.predict(X_test)))

#lr_rmse = rmse(y_test, lr.predict(X_test))
lr_score,lr_rmse
```

# Lasso Algorithm

```python
from sklearn.linear_model import Lasso
lr_lasso = Lasso()
lr_lasso.fit(X_train, y_train)
```

```python
#Check prediction Accuracy
lr_lasso_score = lr_lasso.score(X_test, y_test)

#Root mean Squre Error
lr_lasso_rmse = np.sqrt(mean_squared_error(y_test, lr_lasso.predict(X_test)))
lr_lasso_score, lr_rmse
```

# Ridge Algorithm

```python
from sklearn.linear_model import Ridge
lr_ridge = Ridge()
lr_ridge.fit(X_train, y_train)
```

```python
#Check prediction Accuracy
lr_ridge_score = lr_ridge.score(X_test, y_test)

#Root mean Squre Error
lr_ridge_rmse = np.sqrt(mean_squared_error(y_test, lr_ridge.predict(X_test)))

lr_ridge_score,lr_ridge_rmse
```

# **Random Forest**

```python
# Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)

#Check prediction Accuracy
rfr_score=rfr.score(X_test,y_test) 

#Root mean Squre Error
rfr_rmse = np.sqrt(mean_squared_error(y_test, rfr.predict(X_test)))
rfr_score, rfr_rmse

```

# **Support Vector Machine**

```python

# Support Vector Machine
from sklearn.svm import SVR
svr_model = SVR()
svr_model.fit(X_train, y_train)

#Check prediction Accuracy
svr_score=svr_model.score(X_test,y_test) 

#Root mean Squre Error
svr_rmse = np.sqrt(mean_squared_error(y_test, svr_model.predict(X_test)))
svr_score, svr_rmse

```

# XGBOST

```python
import xgboost
xgb_reg = xgboost.XGBRegressor()
xgb_reg.fit(X_train,y_train)

#Check prediction Accuracy
xgb_reg_score=xgb_reg.score(X_test,y_test) 

#Root mean Squre Error
xgb_reg_rmse = np.sqrt(mean_squared_error(y_test, xgb_reg.predict(X_test)))
xgb_reg_score, xgb_reg_rmse
```

```python
print(pd.DataFrame([{'Model': 'Linear Regression','Score':lr_score, "RMSE":lr_rmse},
              {'Model': 'Lasso','Score':lr_lasso_score, "RMSE":lr_lasso_rmse},
              {'Model': 'Support Vector Machine','Score':svr_score, "RMSE":svr_rmse},
              {'Model': 'Random Forest','Score':rfr_score, "RMSE":rfr_rmse},
              {'Model': 'XGBoost','Score':xgb_reg_score, "RMSE":xgb_reg_rmse}],
             columns=['Model','Score','RMSE']))
```

# **Cross Validation**

```python
# Cross validation
from sklearn.model_selection import cross_val_score
cross_validation = cross_val_score(estimator = rfr, X = X_train,y = y_train, cv = 10)
print("Cross validation accuracy of Random Forest model = ", cross_validation)
print("\nCross validation mean accuracy of Random Forest model = ", cross_validation.mean())
```
